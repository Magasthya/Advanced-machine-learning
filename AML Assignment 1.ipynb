```
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Generating the artificial dataset
centers = [[2, 4], [6, 6], [1, 9]]
n_classes = len(centers)
data, labels = make_blobs(n_samples=150,
                          centers=np.array(centers),
                          random_state=1)

# Splitting the data into training and testing sets (80-20 split)
train_data, test_data, train_labels, test_labels = train_test_split(data, labels,
                                                                    test_size=0.2,
                                                                    random_state=42)

# Performing a KNN analysis
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_data, train_labels)

# Making predictions on the test data
test_predictions = knn.predict(test_data)

# Calculating accuracy
accuracy = accuracy_score(test_labels, test_predictions)
print("Accuracy:", accuracy)

# Plotting the different classes and the decision boundaries
plt.figure(figsize=(10, 6))

# Plotting the training points
for i in range(n_classes):
    plt.scatter(train_data[train_labels == i, 0], train_data[train_labels == i, 1], label=f'Class {i}', alpha=0.8)

# Plotting the test points
for i in range(n_classes):
    plt.scatter(test_data[test_labels == i, 0], test_data[test_labels == i, 1], marker='x', label=f'Test {i}')

# Plotting decision boundaries
h = .02  # step size in the mesh
x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KNN Classification with Simulated Data')
plt.legend()
plt.show()

#It seems like the accuracy of the KNN classifier on the test data is 100%, which means that all test samples were classified correctly.

#This high accuracy suggests that the KNN algorithm performed very well on this artificial dataset with well-separated classes.

```
